```python
# Imlement function that returns plan to stop war given the number of soldiers in each of the two armies and the number of days to stop the war   
# Input: number of soldiers in each of the two armies and the number of days to stop the war
# Output: the number of soldiers in each of the two armies, agressors and defenders, over days 
# Example:
# Input: 200000, 240000, 600  # aggressors, defenders, days

# import the necessary modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
import scipy.stats as stats

# define the war plan function
def war_plan(n_soldiers_1, n_soldiers_2, days_to_stop):

    # define the number of soldiers in each army beginning the war
    n_soldiers_1 = n_soldiers_1
    n_soldiers_2 = n_soldiers_2

    # set number of days to stop the war
    days_to_stop = days_to_stop

    # make a war plan
    # simulate two armies of soldiers fighting over the course of the war
    # define the number of days in the war
    days = np.arange(0, days_to_stop)

    # define the number of soldiers in each army at beginning of the war
    n_soldiers_1 = n_soldiers_1 # agressors
    n_soldiers_2 = n_soldiers_2 # defenders
    ns1 = [n_soldiers_1]
    ns2 = [n_soldiers_2]

    # for every day in the war, calculate the number of soldiers left in each army
    for day in days:

        if n_soldiers_1 > 0 and n_soldiers_2 > 0:
        
            #let the first army lose soldiers as a poissoin process and proportional to the number of soldiers in the second army
            n_soldiers_1 = n_soldiers_1 - np.random.poisson(200)*n_soldiers_2/n_soldiers_1
            #let the second army lose soldiers as a poissoin process and proportional to the number of soldiers in the first army
            n_soldiers_2 = n_soldiers_2 - np.random.poisson(100)*n_soldiers_1/n_soldiers_2

            ns1.append(n_soldiers_1)
            ns2.append(n_soldiers_2) 
            # print the number of soldiers in each army at the end of the war
            print("After day", day, "there are", n_soldiers_1, "soldiers in the agressor army and", n_soldiers_2, "soldiers in defending army")

    return ns1, ns2
```

```python
a1, a2 = war_plan(200000, 240000, 600)
```

```python
#plot the number of soldiers in each army over the course of the war
plt.plot(a1, label = "Aggresor Army")
plt.plot(a2, label = "Defending Army")
# label the x-axis
plt.xlabel("Days")
# label the y-axis
plt.ylabel("Number of Soldiers")
# label the title
plt.title("Number of Soldiers in Each Army Over Time")
plt.legend()
plt.show()
```
